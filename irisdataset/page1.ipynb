```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
iris = pd.read_csv("irisdataset.csv")
iris.shape
iris.columns
iris.value_counts()
```

```python
iris.plot(kind="scatter",grid=True,x="sepal_length",y="sepal_width")
iris.plot(kind="scatter",x="petal_length",y="petal_width")
plt.tight_layout
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(iris,hue="species",height=5).map(plt.scatter,"sepal_length","sepal_width").add_legend()
# Linearly seperated
```

### Obeservations:

1. Using sepal_length and sepal_width we can distinguish setosa flowers from others.
2. Seperating versicolor and virginica is hard, as there is much overlapping in them.
